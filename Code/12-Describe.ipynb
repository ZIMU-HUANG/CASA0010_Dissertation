```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset, assuming the file name is 'Dataset.csv'
data = pd.read_csv('Data/Final/Dataset.csv')

# View the first few rows of the data
data.head()
```

```python
# Define a custom format function to display numbers
def custom_format(x):
    return f'{x:.2f}'.rstrip('0').rstrip('.') if isinstance(x, float) else x

# Perform basic descriptive statistical analysis
data_summary = data.describe()

# Apply the custom format and table styling
styled_data = data_summary.style.format(custom_format).set_table_styles([
    {
        'selector': 'th',
        'props': [('font-size', '8px')]
    },
    {
        'selector': 'tbody td',
        'props': [('font-size', '8px')]
    }
])

# Display the styled DataFrame
styled_data
```

```python
import matplotlib.pyplot as plt

# Scatter plot of IMD_Change vs. Health_Change
plt.figure(figsize=(10, 6))
plt.scatter(data['IMD_Change'], data['Health_Change'], alpha=0.5, c=data['Count'], cmap='viridis')
plt.colorbar(label='Count')
plt.title('Scatter Plot of IMD Change vs. Health Change')
plt.xlabel('IMD Change')
plt.ylabel('Health Change')
plt.grid(True, linestyle='--', alpha=0.6)

# Save the figure as an image
output_path = "Data/scatter plot.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```

```python
# Correlation matrix
correlation_matrix = data[['IMD_Change', 'Health_Change', 'Count']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of IMD Change, Health Change, and Count')

# Save the figure as an image
output_path = "Data/CM.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```

```python
# 1. Histogram: Distribution of IMD_Change
plt.figure(figsize=(12, 6))
plt.hist(data['IMD_Change'], bins=20, color='#b30425', edgecolor='black')
plt.title('Distribution of IMD Change (2020 - 2021)')
plt.xlabel('IMD Change')
plt.ylabel('Frequency')

# Save the figure as an image
output_path = "Data/Distribution of IMD Change (2020 - 2021).png"
plt.savefig(output_path, dpi=300, bbox_inches='tight', transparent=True)

plt.show()

# 2. Histogram: Distribution of Health_Change
plt.figure(figsize=(12, 6))
plt.hist(data['Health_Change'], bins=20, color='#3b4cc0', edgecolor='black')
plt.title('Distribution of Health Change (2020 - 2021)')
plt.xlabel('Health Change')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the figure as an image
output_path = "Data/Distribution of Health Change (2020 - 2021).png"
plt.savefig(output_path, dpi=300, bbox_inches='tight', transparent=True)

plt.show()

# 3. Box Plot: IMD_Change
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.boxplot(data['IMD_Change'], vert=False)
plt.title('Box Plot of IMD Change')
plt.xlabel('IMD Change')

# 4. Box Plot: Health_Change
plt.subplot(1, 2, 2)
plt.boxplot(data['Health_Change'], vert=False)
plt.title('Box Plot of Health Change')
plt.xlabel('Health Change')

plt.tight_layout()

# Save the figure as an image
output_path = "Data/Box.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight', transparent=True)

plt.show()
```
